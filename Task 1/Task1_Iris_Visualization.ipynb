## Dataset Understanding

The Iris dataset contains measurements of iris flowers including sepal length, sepal width, petal length, petal width, and species. Before visualization, it is important to inspect the data structure and verify whether missing values are present.

```python
# Check duplicate rows

df.duplicated().sum()
```

```python
# Remove unnecessary Id column

df = df.drop("Id", axis=1)
df.head()
```

## Data Cleaning

The dataset was checked for missing values and duplicate records. No significant data quality issues were found. If present, the unnecessary `Id` column was removed because it does not contribute to the analysis.

```python
# Scatter Plot

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='SepalLengthCm',
    y='PetalLengthCm',
    hue='Species',
    s=80
)

plt.title("Sepal Length vs Petal Length")
plt.show()
```

### Observation

The scatter plot shows a clear relationship between sepal length and petal length. The Setosa species forms a distinct cluster, while Versicolor and Virginica have some overlap.

```python
# Histogram

plt.figure(figsize=(8,6))

sns.histplot(df['SepalLengthCm'], bins=15, kde=True)

plt.title("Distribution of Sepal Length")

plt.show()
```

### Observation

The histogram illustrates the distribution of sepal length values. Most flowers have sepal lengths between 5 and 7 centimeters, indicating a moderately symmetric distribution.

```python
# Box Plot

plt.figure(figsize=(8,6))

sns.boxplot(
    data=df,
    x='Species',
    y='PetalLengthCm'
)

plt.title("Petal Length by Species")

plt.show()
```

### Observation

The box plot highlights differences in petal length among the three species. Setosa has the smallest petal lengths, Virginica has the largest, and Versicolor falls in between. A few potential outliers are also visible.

```python
# Pair Plot

sns.pairplot(df, hue='Species')

plt.show()
```

```python
# Correlation Heatmap

plt.figure(figsize=(8,6))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm',
            linewidths=0.5)

plt.title("Correlation Heatmap")

plt.show()
```
