```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset (built-in)
df = sns.load_dataset('iris')

# Display basic info
print(df.shape)
print(df.columns)
print(df.head())
```

```python
# Scatter plot - relationship between sepal length and petal length
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')
plt.title('Sepal Length vs Petal Length')
plt.show()
```

```python
# Histogram - distribution of sepal length
df['sepal_length'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Count')
plt.show()
```

```python
# Box plot - spread and outliers of each species
sns.boxplot(data=df, x='species', y='sepal_length')
plt.title('Sepal Length by Species')
plt.show()
```

```python
# Conclusion
print("""
Key Insights from Iris Dataset:
1. Virginica has the largest sepal and petal sizes
2. Setosa is clearly separated from other species
3. Petal length increases as sepal length increases
4. One outlier found in Virginica species
""")
```
